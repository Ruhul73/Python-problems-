Find highest price
find lowest price
find the range of price

```python
cars = {"alto": 1800000, "city": 2500000, "santro": 1200000, "sportage": 4300000, "cultus": 2300000}

# Find the car with the highest price
highest_priced_car = max(cars, key=cars.get)
print("Car with the highest price:", highest_priced_car)

# Find the car with the lowest price
lowest_priced_car = min(cars, key=cars.get)
print("Car with the lowest price:", lowest_priced_car)

# Find the cars with a price range of 1500000-2500000
cars_range = []
for car, price in cars.items():
    if 1500000 <= price <= 2500000:
        cars_range.append(car)

print("Cars with a price range of 1500000-2500000:", cars_range)

```

note counter machine

```python
amount = int(input("Enter the amount to dispense: "))

notes_5000 = amount // 5000
remaining_amount = amount % 5000

notes_1000 = remaining_amount // 1000
remaining_amount = remaining_amount % 1000

notes_500 = remaining_amount // 500

print(f"Number of 5000 notes: {notes_5000}")
print(f"Number of 1000 notes: {notes_1000}")
print(f"Number of 500 notes: {notes_500}")

if remaining_amount % 500 != 0:
    print(f"Remaining amount (less than 500): {remaining_amount % 500}")
```
